```python
import pandas as pd
data = pd.read_csv('regrex1.csv')
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot
plt.scatter(data['x'], data['y'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of x vs y')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(data[['x']], data['y'])
data['y_pred'] = model.predict(data[['x']])
```

```python
# Plot the original data and the linear model
plt.scatter(data['x'], data['y'], label='Original data')
plt.plot(data['x'], data['y_pred'], color='pink', label='Linear model')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of x vs y with Linear Model')
plt.legend()
plt.show()
```
